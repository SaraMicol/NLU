IMPORTING THE LIBRARIES:

```python
from data_utils import read_file
from conllu_reader import ConlluReader
from model import ParserMLP
from algorithm import ArcEager
from data_utils import update_trees
```

```python
# Initialize the ConlluReader
reader = ConlluReader()
test_trees = read_file(reader,path="en_partut-ud-test_clean.conllu", inference=True)
train_trees = read_file(reader,path="en_partut-ud-train_clean.conllu", inference=False)
dev_trees = read_file(reader,path="en_partut-ud-dev_clean.conllu", inference=False)
print(test_trees[0])
```

PREPARE SAMPLE FOR TRAINING:

```python
arc_eager = ArcEager()
train_all_samples = []
for tree in train_trees:
    samples = arc_eager.oracle(tree)
    for sample in samples:
        train_all_samples.append(sample)

dev_all_samples = []
for tree in dev_trees: 
    samples = arc_eager.oracle(tree)
    for sample in samples:
        dev_all_samples.append(sample)
```

TRAIN THE MODEL:

```python
arc_eager=ArcEager()
parser_mlp1.train(train_all_samples,dev_all_samples,arc_eager)
#parser_mlp2.train(train_all_samples,dev_all_samples,arc_eager)
#parser_mlp3.train(train_all_samples,dev_all_samples,arc_eager)
#parser_mlp4.train(train_all_samples,dev_all_samples,arc_eager)
#parser_mlp5.train(train_all_samples,dev_all_samples,arc_eager)
```

HORIZONTAL PREDICTION:

```python
n = 50  
sentences= test_trees[:n]

final_states=parser_mlp1.run(sentences,arc_eager)
```

UPDATE THE TREES

```python
updated_trees = update_trees(test_trees, final_states)

for i, tree in enumerate(updated_trees, 1):
    print(f"\n=== Tree {i} ===")
    for token in tree:
        print(token)
```

POSTPROCESSING:

```python
reader = ConlluReader()

output_filename = "updated_trees.conllu"

with open(output_filename, "w", encoding="utf-8") as conllu_file:
    for i, tree in enumerate(updated_trees, 1):
        # Converte l'albero in formato stringa Conllu
        conllu_str = reader.tree2conllustr(tree)
        # Scrive ogni albero nel file separandoli con una riga vuota
        conllu_file.write(f"# Tree {i}\n{conllu_str}\n\n")

print(f"File saved as {output_filename}")

```

```python
from postprocessor import PostProcessor
from conllu_reader import ConlluReader  # Ensure this is the correct library for reading .conllu files

# Initialize the post-processor
p = PostProcessor()

# Read and display the file before postprocessing
reader = ConlluReader()
print("An example of a corrupted tree, before postprocessing:")
trees = reader.read_conllu_file("updated_trees.conllu")
for tree in trees:
    for token in tree:
        print(token)

print("\n\n")

# Perform postprocessing
trees = p.postprocess("updated_trees.conllu")
print("An example of a (now no) corrupted tree, after postprocessing:")
for tree in trees:
    for token in tree:
        print(token)

# Save the postprocessed trees to a new file
output_file = "postprocessed_trees.conllu"
with open(output_file, "w", encoding="utf-8") as f:
    for tree in trees:
        for token in tree:
            f.write(f"{token}\n")  # Write each token to the file
        f.write("\n")  # Add a blank line to separate trees
print(f"Postprocessed trees saved to {output_file}")
```
